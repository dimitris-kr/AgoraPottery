```python
# Local Path for Running Locally
path = "../data"
```

```python
# File path for the text file
file_path = path + 'Agora12.2024.10.25.txt'

# List to hold each parsed record
records = []

# Dictionary to hold each record's fields
record = {}
```

```python
# Function to clean up field values (remove HTML tags, extra spaces)
def clean_value(value):
    # Remove HTML tags and excess whitespace
    return re.sub(r'<.*?>', '', value).strip()
```

```python
# Read file and parse line by line
with open(file_path, 'r', encoding='utf-8') as file:
    for line in file:
        # Detect new record by "Id : Agora:Object:Agora"
        if re.match(r'^Id\s+: Agora:Object:Agora', line):
            # Save the previous record if it exists
            if record:
                records.append(record)
                record = {}

        # Identify fields using `: ` separator
        if ': ' in line:
            field, value = line.split(': ', 1)
            record[field.strip()] = clean_value(value)

    # Append last record if file ends without extra newline
    if record:
        records.append(record)
```

```python
# Convert records list to DataFrame
df = pd.DataFrame(records)
df
```

```python
# Save DataFrame to CSV
df.to_csv(path + 'agora12_data.csv', index=False)
print("Data saved to agora12_data.csv")
```
